# Data Analysis
## De'Aaron Fox Scoring Splits (Shot Tendencies)
Steps for this project
- load in csv
- clean up the data and only keep the columns needed
- calculate total average for all seasons
- plot on pie chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fox=pd.read_csv(r"/Users/ryantseng/Documents/De'Aaron Fox Scoring Splits - Sheet1.csv")
#print(fox)
```

```python
##new dataset with only necessary columns
cleanfox=fox[['%PTS 2PT MR','%PTS 3PT','%PTS FT','%PTS PITP']]
print(cleanfox)
```

```python
#check if all values equal 100
check = cleanfox.sum(axis=1)

for i in range(len(cleanfox.columns)):
    
    if check[i] < 99.0 or check[i]>100.0:
       print("error")
    
#Find totals
total = cleanfox.sum(axis=0)/len(cleanfox.index)
```

```python
#plot results onto pie chart
labels = 'Mid-Range Jumpers', '3-Pointers','Free-Throws','Points in the Paint'
sizes = total
explode = (0,0.1,0,0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("De'Aaron Fox's FGM Shot Tendency")

fig = plt.gcf()
fig.set_size_inches(7,7) 

plt.show()
```
