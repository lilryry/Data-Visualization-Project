# Data Analysis
## Oakland Gunshot Statistics   2011-2016

Steps for this project
- import csv files
- only keep data related to gunshot incidents
- count number of incidents per year
- graph number of incidents vs year line graph

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds2011=pd.read_csv(r"/Users/ryantseng/Documents/Oakland Statistics/records-for-2011.csv")
ds2012=pd.read_csv(r"/Users/ryantseng/Documents/Oakland Statistics/records-for-2012.csv")
ds2013=pd.read_csv(r"/Users/ryantseng/Documents/Oakland Statistics/records-for-2013.csv")
ds2014=pd.read_csv(r"/Users/ryantseng/Documents/Oakland Statistics/records-for-2014.csv")
ds2015=pd.read_csv(r"/Users/ryantseng/Documents/Oakland Statistics/records-for-2015.csv")
ds2016=pd.read_csv(r"/Users/ryantseng/Documents/Oakland Statistics/records-for-2016.csv")


```

```python
#count the number of "gunshot" related incidences each year
count2011=ds2011['Incident Type Description'].str.count("GUNSHOTS").sum()
count2012=ds2012['Incident Type Description'].str.count("GUNSHOTS").sum()
count2013=ds2013['Incident Type Description'].str.count("GUNSHOTS").sum()
count2014=ds2014['Incident Type Description'].str.count("GUNSHOTS").sum()
count2015=ds2015['Incident Type Description'].str.count("GUNSHOTS").sum()
count2016=ds2016['Incident Type Description'].str.count("GUNSHOTS").sum()
```

```python
#begin graphing
year = [2011,2012,2013,2014,2015,2016]
gunshots = [count2011,count2012,count2013,count2014,count2015,count2016]

plt.plot(year, gunshots)
plt.title('Gun-related crime in Oakland, CA')
plt.xlabel('Year')
plt.ylabel('# of Gunshot Incidences')
plt.show()
```
